# Visualization Data

```julia
using Plots
```

```julia
x = 1:15; y = rand(15)
```

```julia
plot(x, y)
```

```julia
z = rand(15);
```

```julia
# y will be preserved and z will be added in the plot
plot!(x, z)
```

```julia
# same as previous
p = plot(x, y)
plot!(p, x, z)
```

```julia
plot(x,y, title="random title")
```

```julia
plot(x, y, title="random plot", label="line chat(label)")
```

```julia
plot(x, y, xlabel = "X axis label", ylabel="Y axis label")
```

```julia
scatter(x,y)
```

```julia
bar(x,y)
```

```julia
histogram(y)
```

```julia
p1 = plot(x, y)
p2 = scatter(x,y)
p3 = bar(x,y)
p4 = histogram(y)
# can change the title and many things in this beautiful rep.
plot(p1,p2,p3,p4, layout=(2,2), legend = false)
```

```julia
y = rand(15 ,4)

# plot bw a 1D vector and 2D vector.
# 1D is the x axis and every column of 2D becomes y axis.
# we can have any charts with this representation.
plot(x, y, layout=(4,1))
```
